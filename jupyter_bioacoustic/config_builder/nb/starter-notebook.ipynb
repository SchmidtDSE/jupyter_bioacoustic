### Data

The `data` parameter tells the annotator where your rows come from. The source type is auto-detected:

| You pass | Detected as | Example |
|---|---|---|
| Path to a local file | File (CSV, Parquet, JSONL) | `"detections.csv"` |
| A URL | Remote file | `"https://example.com/data.csv"` |
| A string with `SELECT` | SQL query ([DuckDB](https://duckdb.org/)) | `"SELECT * FROM 'data.parquet' WHERE conf > 0.5"` |
| `api::` prefix | API endpoint | `"api::https://api.example.com/detections"` |
| A DataFrame | In-memory data | `my_dataframe` |

You can also pass a **dict** for explicit control over the source, columns, and secrets:

```python
data={'path': 'data.csv', 'columns': ['species', 'confidence', 'start_time', 'end_time']}
```

> **Docs:** [The Data Parameter](https://schmidtdse.github.io/jupyter_bioacoustic/sections/data.html)

---

### Audio

The `audio` parameter specifies where audio files live. Also auto-detected:

| You pass | Detected as | Example |
|---|---|---|
| A file path | Local file | `"recordings/site-a.flac"` |
| An `https://` URL | Remote file (downloaded & cached) | `"https://example.com/rec.flac"` |
| An `s3://` URI | S3 (partial byte-range fetch) | `"s3://bucket/recording.flac"` |
| A column name (no `/` or `.`) | Per-row column mode | `"audio_path"` |

Use `audio_prefix` and `audio_suffix` to build paths from column values. For example, if your data has a `filename` column:

```python
audio='filename', audio_prefix='audio/', audio_suffix='.flac'
```

> **Docs:** [The Audio Parameter](https://schmidtdse.github.io/jupyter_bioacoustic/sections/audio.html)

### Config Files (Project / Config / Form)

Any constructor parameter can also be set in a **YAML or JSON config file**. Config files are recommended for anything beyond simple setups — they keep notebooks clean and are easy to share.

You can split settings across up to three files:

| File | Purpose |
|---|---|
| **Project file** | Shared settings: data source, audio source, output path |
| **Config file** | App behavior: columns, spectrogram capture, layout |
| **Form file** | Form elements: selects, textboxes, checkboxes, dynamic forms |

```python
ba = BioacousticAnnotator(
    project='project.yaml',
    config='config.yaml',
    form_config='form.yaml',
)
ba.open()
```

Or use a single file with everything:

```python
ba = BioacousticAnnotator(config='my-config.yaml')
ba.open()
```

> **Docs:** [Config Files & Usage](https://schmidtdse.github.io/jupyter_bioacoustic/sections/usage.html)
